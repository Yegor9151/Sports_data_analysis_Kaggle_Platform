## libs

```python
%%time
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## get data

```python
cb_cv_cat = pd.read_csv('./cb_cv_cat.csv')
cb_cv_cat_def = pd.read_csv('./cb_cv_cat_def.csv')
cb_cv_dim = pd.read_csv('./cb_cv_dim.csv')

lgbm_cv_cat = pd.read_csv('./lgbm_cv_cat.csv')
lgbm_cv_cat_def = pd.read_csv('./lgbm_cv_cat_def.csv')
lgbm_cv_dim = pd.read_csv('./lgbm_cv_dim.csv')

xgb_cv_cat = pd.read_csv('./xgb_cv_cat.csv')
xgb_cv_dim = pd.read_csv('./xgb_cv_dim.csv')
```

# Задание 10: построить ROC-кривую для всех построенных алгоритмов на обучающей и тестовой выборке. Сделать выводы о работе алгоритмов с точки зрения качества на тестовой выборке и с точки зрения переобучения.

```python
plt.figure(figsize=(16, 8))
plt.title('ROC-AUC plot for all algorithms')

cb_cv_dim['test-AUC-mean'].plot(label='cb_cv_dim', style='-')
cb_cv_cat['test-AUC-mean'].plot(label='cb_cv_cat', style='-')
cb_cv_cat_def['test-AUC-mean'].plot(label='cb_cv_cat_def', style='-')

lgbm_cv_dim['auc-mean'].plot(label='lgbm_cv_dim', style='--')
lgbm_cv_cat['auc-mean'].plot(label='lgbm_cv_cat', style='--')
lgbm_cv_cat_def['auc-mean'][:200].plot(label='lgbm_cv_cat_def', style='--')

xgb_cv_cat['train-auc-mean'].plot(label='xgb_cv_cat - train', style='-.')
xgb_cv_cat['test-auc-mean'].plot(label='xgb_cv_cat - test', style='-.')
xgb_cv_dim['train-auc-mean'].plot(label='xgb_cv_dim - train', style='-.')
xgb_cv_dim['test-auc-mean'].plot(label='xgb_cv_dim - test', style='-.')

plt.xlabel('iteration')
plt.ylabel('ROC-AUC')
plt.legend()
```

## Вывод:
- Классический XGB показывает более плавную ROC-AUC кривую и более высокие показатели на меньшее число итераций.
- У LGBM и CB показывают более очевидную тенденцию к повышению ROC-AUC кривой.
